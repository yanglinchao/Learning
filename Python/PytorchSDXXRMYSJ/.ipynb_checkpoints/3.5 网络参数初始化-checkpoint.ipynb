```python
import torch.nn as nn
# 
class TestNet(nn.Module):
    def __init__(self):
        super(TestNet, self).__init__()
        self.conv1 = nn.Conv2d(3, 16, 3)
        self.hidden = nn.Sequential(
            nn.Linear(100, 100),
            nn.ReLU(),
            nn.Linear(100, 50),
            nn.ReLU()
        )
        self.cla = nn.Linear(50, 10)
    # 
    def forward(self, x):
        x = self.conv1(x)
        x = x.view(x.shape[0], -1)
        x = self.hidden(x)
        output = self.cla(x)
        return output
# 
testnet = TestNet()
print(testnet)
```

> init_weights()

```python
# 
def init_weights(m):
    # 
    if type(m) == nn.Conv2d:
        torch.nn.init.normal(m.weight, mean=0, std=0.5)
    # 
    if type(m) == nn.Linear:
        torch.nn.init.uniform(m.weight, a=-0.1, b=0.1)
        m.bias.data.fill_(0.01)
```

> typenn.Conv2d00.5weight(-0.1, 0.1)weightbias0.01<p> testnetinit_weights()applytestnet

```python
# apply
torch.manual_seed(13) # 
testnet.apply(init_weights)
```
