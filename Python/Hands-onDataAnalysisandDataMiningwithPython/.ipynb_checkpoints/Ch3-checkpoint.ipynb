# 3.1 数据质量分析

### 使用describe()方法查看数据的基本情况

```python
import pandas as pd
working_dir = "C:/Users/ylc/GitHub/Learning/Python/Hands-onDataAnalysisandDataMiningwithPython"
catering_sale = working_dir + "/catering_sale.xls"
data = pd.read_excel(catering_sale, index_col='日期') # 读取数据，制定“日期”列为索引列
print(data.describe()) # 使用describe()方法查看数据的基本情况情况
```

> 其中count是非空值数，通过len(data)可以知道数据记录为201条，因此缺失值数量为1

### 使用箱图进行异常值检测

```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  #用来显示正常负号

plt.figure() # 建立图像
p = data.boxplot(return_type='dict') # 画箱型图，直接使用DataFrame的方法
x = p['fliers'][0].get_xdata() # 'flies'即为异常值的标签
y = p['fliers'][0].get_ydata()
y.sort()
'''
用annotate添加注释
%qtconsole中有些相近的点，注释会出现重叠
以下参数都是经过调试的，需要具体问题具体调试
'''
for i in range(len(x)):
    if i>0:
        plt.annotate(y[i], xy=(x[i], y[i]), xytext=(x[i]+0.05 -0.8/(y[i]-y[i-1]), y[i]))
    else:
        plt.annotate(y[i], xy=(x[i], y[i]), xytext=(x[i]+0.08, y[i]))
plt.show()
```

## 3.2.5 贡献度分析

### 帕累托图

```python
import pandas as pd

# 初始化参数
working_dir = "C:/Users/ylc/GitHub/Learning/Python/Hands-onDataAnalysisandDataMiningwithPython"
dish_profit = working_dir + "/catering_dish_profit.xls"
data = pd.read_excel(dish_profit, index_col="菜品名")
data = data['盈利'].copy()
data.sort_values(ascending=False)

import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False # 用来正常显示负号
plt.figure()
data.plot(kind = 'bar')
plt.ylabel('盈利（元）')
p = 1.0*data.cumsum()/data.sum()
p.plot(color='r', secondary_y=True, style='-o', linewidth=2)
plt.annotate(format(p[6], '.4%'), xy=(6, p[6]), xytext=(6*0.9, p[6]*0.9), arrowprops=dict(arrowstyle="->", connectionstyle="arc3,rad=.2"))
# 添加注释，即85%处的标记。这里包括了指定箭头样式
plt.ylabel('盈利（比例）')
plt.show()
```

### 相关系数

#### Peason相关系数

```python
import pandas as pd
working_dir = "C:/Users/ylc/GitHub/Learning/Python/Hands-onDataAnalysisandDataMiningwithPython"
catering_sale = "/catering_sale_all.xls"
data = pd.read_excel(working_dir+catering_sale, index_col="日期")

print(data.corr(method='pearson')) # Pearson相关系数矩阵，给出了任意两款菜式之间的相关系数
```

#### Spearman相关系数

```python
print(data.corr(method='spearman')['百合酱蒸凤爪']) # Spearman相关系数
```

```python
print(data['百合酱蒸凤爪'].corr(data['翡翠蒸香茜饺'], method='spearman'))
```

Pearson相关系数指线性相关，在计算Pearson相关系数前要先进行正态性检验<br>
Spearman相关系数指秩相关
