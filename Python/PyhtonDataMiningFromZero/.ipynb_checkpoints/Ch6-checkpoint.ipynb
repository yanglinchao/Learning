### 纵向条形图

```python
import pandas as pd
import matplotlib.pyplot as plt

# 读入数据
GDP = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/Province GDP 2017.xlsx")
# 设置绘图风格
plt.style.use('ggplot') # 使用R语言中的ggplot2风格
# 绘制条形图
plt.bar(x=range(GDP.shape[0]), # 横坐标
        height=GDP.GDP, # 指定条形图y轴的数值
        tick_label = GDP.Province, # 指定条形图x轴的刻度标签
        color = 'steelblue' # 指定条形图的填充色
       )
plt.rcParams['font.sans-serif'] = 'SimHei'
plt.ylabel('GDP(万亿)') # 添加y轴的标签
plt.title('2017年度6个省份GDP分布') # 添加条形图的标题
# 为每个条形图添加数值标签
for x, y in enumerate(GDP.GDP):
    plt.text(x, y+0.1, '%s' %round(y, 1), ha='center')
plt.show() # 显示图形
```

### 堆叠条形图

```python
import pandas as pd
import matplotlib.pyplot as plt
Industry_GDP = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/Industry_GDP.xlsx")

Quarters = Industry_GDP.Quarter.unique() # 取出四个不同的季度标签，用作堆叠条形图x轴的刻度标签
Industry1 = Industry_GDP.GDP[Industry_GDP.Industry_Type == "第一产业"] # 取出第一产业的四季度值
Industry1.index = range(len(Quarters)) # 重新设置行索引
Industry2 = Industry_GDP.GDP[Industry_GDP.Industry_Type == "第二产业"] # 取出第二产业的四季度值
Industry2.index = range(len(Quarters)) # 重新设置行索引
Industry3 = Industry_GDP.GDP[Industry_GDP.Industry_Type == "第三产业"] # 取出第三产业的四季度值

# 绘制堆叠条形图
plt.bar(x = range(len(Quarters)), height=Industry1, color='steelblue', label='第一产业', tick_label=Quarters) # 各季度第一产业条形图
plt.bar(x = range(len(Quarters)), height=Industry2, bottom=Industry1, color='green', label='第二产业', tick_label=Quarters) # 各季度第二产业条形图
plt.bar(x = range(len(Quarters)), height=Industry3, bottom=Industry1+Industry2, color='red', label='第三产业', tick_label=Quarters) # 各季度第三产业条形图
plt.ylabel("生产总值（亿）")
plt.title("2017年各季度三产业总值")
plt.legend() # 显示图例
plt.show()
```

### 簇状条形图

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

HuRun = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/HuRun.xlsx")
Cities = HuRun.City.unique() # 取出城市名
Counts2016 = HuRun.Counts[HuRun.Year == 2016] # 取出2016年各城市亿万资产家庭数
Counts2017 = HuRun.Counts[HuRun.Year == 2017] # 取出2017年各城市亿万资产家庭数

# 绘制簇状条形图
bar_width = 0.4 # 设置宽度
plt.bar(x=np.arange(len(Cities)), height=Counts2016, label='2016', color='steelblue', width=bar_width)
plt.bar(x=np.arange(len(Cities))+bar_width, height=Counts2017, label='2017', color='indianred', width=bar_width) # 利用偏移条形宽度画图
plt.xticks(np.arange(5)+0.2, Cities) # 将刻度标签偏移
plt.legend()
plt.show()
```

### 箱线图

```python
import pandas as pd
import matplotlib.pyplot as plt
Sec_Buildings = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/sec_buildings.xlsx")
plt.boxplot(x=Sec_Buildings.price_unit,
            patch_artist=True, # 要求用自定义颜色填充，默认白色填充
            boxprops={'color':'black', 'facecolor':'steelblue'}, # 设置箱体属性，如边框色和填充色
            flierprops={'marker':'o', 'markerfacecolor':'red', 'markersize':3}, # 设置异常点的属性，如点的形状、填充色和大小
            meanprops={'marker':'D', 'markerfacecolor':'indianred', 'markersize':4}, # 设置均值点的属性，如点的形状、填充色和大小
            medianprops={'linestyle':'--', 'color':'orange'}, # 设置中位数线的属性，如线的类型和颜色
            labels={''} # 删除x轴的刻度标签，否则图形显示刻度标签为1
           )
plt.title('二手房单价分布的箱线图')
plt.show()
```

### 分组箱线图

```python
import pandas as pd
import matplotlib.pyplot as plt
Sec_Buildings = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/sec_buildings.xlsx")
group_region = Sec_Buildings.groupby('region')
avg_price = group_region.aggregate({'price_unit':np.mean}).sort_values('price_unit', ascending=False)

# 通过循环，将不同行政区域的二手房存储到列表中
region_price = []
for region in avg_price.index:
    region_price.append(Sec_Buildings.price_unit[Sec_Buildings.region == region])
# 绘制分组箱线图
plt.boxplot(x=region_price,
            patch_artist=True,
            labels=avg_price.index, # 添加x轴的刻度标签
            showmeans=True,
            boxprops={'color':'black', 'facecolor':'steelblue'},
            flierprops={'marker':'o', 'markerfacecolor':'red', 'markersize':3},
            meanprops={'marker':'D', 'markerfacecolor':'indianred', 'markersize':4},
            medianprops={'linestyle':'--', 'color':'orange'})
plt.ylabel('单价（元）')
plt.title('不同行政区域的二手房单价对比')
plt.show()
```

### 两条折线图

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
wechat = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/wechat.xlsx")

# 绘制阅读人数折线图
plt.plot(wechat.Date, # x轴数据
        wechat.Counts, # y轴数据
        linestyle = '-', # 折线类型，虚线
        color = 'steelblue', # 折线颜色
        label = '阅读人次'
        )
plt.plot(wechat.Date, # x轴数据
        wechat.Times, # y轴数据
        linestyle = '--', # 折线类型，虚线
        color = 'indianred', # 折线颜色
        label = '阅读人次'
        )
ax = plt.gca() # 获取图的坐标信息
date_format = mpl.dates.DateFormatter("%m-%d") # 设置日期的显示格式
ax.xaxis.set_major_formatter(date_format)
xlocator = mpl.ticker.MultipleLocator(7) # 设置x轴每个刻度的间隔天数
ax.xaxis.set_major_locator(xlocator)
plt.xticks(rotation=45) # 为了避免x轴刻度标签的紧凑，将刻度标签旋转45度
plt.ylabel('人数')
plt.title('每天微信文章阅读人数与人次趋势')
plt.legend()
plt.show()
```

### 气泡图

```python
import pandas as pd
import matplotlib.pyplot as plt
Prod_Category = pd.read_excel("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/SuperMarket.xlsx")

# 将利润率标准化到[0, 1]之间（因为利润率中有负数），然后加上微小的数值0.001
range_diff = Prod_Category.Profit_Ratio.max()-Prod_Category.Profit_Ratio.min()
Prod_Category['std_ratio'] = (Prod_Category.Profit_Ratio-Prod_Category.Profit_Ratio.min())/range_diff + 0.001

# 绘制办公用品的气泡图
type = ['办公用品', '技术产品', '家具产品']
color = ['steelblue', 'indianred', 'black']
for i in range(0, 2):
    plt.scatter(x=Prod_Category.Sales[Prod_Category.Category == type[i]],
           y=Prod_Category.Profit[Prod_Category.Category == type[i]],
           s=Prod_Category.std_ratio[Prod_Category.Category == type[i]]*500,
           color=color[i], label=type[i], alpha=0.6)
plt.xlabel('销售额')
plt.ylabel('利润')
plt.title('销售额、利润及利润率的气泡图')
plt.legend()
plt.show()
```
