## CART算法

### 离散变量

#### 数据准备

```python
import pandas as pd
Titanic = pd.read_csv("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/Titanic.csv")
Titanic.head()
```

```python
Titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True) # 删除不使用的变量
Titanic.isnull().sum(axis=0) # 查看缺失值
```

```python
# 对Sex分组，用各组乘客的平均年龄填充各组中的缺失年龄
fillna_Titanic = []
for i in Titanic.Sex.unique():
    update = Titanic.loc[Titanic.Sex==i, ].fillna(value={'Age':Titanic.Age[Titanic.Sex==i].mean()})
    fillna_Titanic.append(update)
Titanic = pd.concat(fillna_Titanic)
# 使用Embarked变量的众数填充缺失值
Titanic.fillna(value={'Embarked':Titanic.Embarked.mode()[0]}, inplace=True)
# 将数值型的Pclass转换为类别型，否则无法对其哑变量处理
Titanic.Pclass = Titanic.Pclass.astype('category')
dummy = pd.get_dummies(Titanic[['Sex', 'Embarked', 'Pclass']]) # 哑变量处理
Titanic = pd.concat([Titanic, dummy], axis=1)
# 删除原始的Sex、Embarked和Pclass变量
Titanic.drop(['Sex', 'Embarked', 'Pclass'], inplace=True, axis=1)
Titanic.head()
```

#### 构建决策树模型

```python
from sklearn import model_selection
predictors = Titanic.columns[1:] # 取出所有自变量名称
# 将数据集拆分为训练集和测试集，且测试集比例为25%
X_train, X_test, y_train, y_test = model_selection.train_test_split(Titanic[predictors], Titanic.Survived,
                                                                    test_size=0.25, random_state=1234)
```

> 为了防止构建的决策树产生过拟合，需要对决策树进行预剪枝，如限制树生长的最大深度、设置决策树的中间节点能够继续分支的最小样本量以及叶节点的最小样本量等。Python提供了网格搜索法，可以帮助用户快速地进行各参数组合下的试错，网格搜索法的实现需要调用GridSearchCV类，该“类”存储在sklearn的子模块model_selection中

```python
from sklearn.model_selection import GridSearchCV
from sklearn import tree
# 预设各参数的不同选项值
max_depth = [2, 3, 4, 5, 6]
min_samples_split = [2, 4, 6, 8]
min_samples_leaf = [2, 4, 8, 10, 12]
parameters = {'max_depth':max_depth, 'min_samples_split':min_samples_split, 'min_samples_leaf':min_samples_leaf}
# 网格搜索法，测试不同的参数值
grid_dtcateg = GridSearchCV(estimator=tree.DecisionTreeClassifier(), param_grid=parameters, cv=10)
grid_dtcateg.fit(X_train, y_train) # 模型拟合
grid_dtcateg.best_params_ # 返回最佳组合的参数值
```

```python
# 构建分类决策树
CART_Class = tree.DecisionTreeClassifier(max_depth=6, min_samples_leaf=8, min_samples_split=6)
decision_tree = CART_Class.fit(X_train, y_train) # 模型拟合
pred = CART_Class.predict(X_test) # 模型在测试集上的预测
```

#### 模型效果评价

```python
import matplotlib.pyplot as plt
from sklearn import metrics
y_score = CART_Class.predict_proba(X_test)[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, y_score)
roc_auc = metrics.auc(fpr, tpr)
plt.stackplot(fpr, tpr, color='steelblue', alpha=0.5, edgecolor='black') # 绘制面积图
plt.plot(fpr, tpr, color='black', lw=1)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.text(0.5, 0.3, 'ROC curve (area=%0.2f)' % roc_auc)
plt.xlabel('1-Specificity')
plt.ylabel('Sensitivity')
plt.show()
```

#### 绘制决策树

```python
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
from sklearn.externals.six import StringIO
# 绘制决策树
dot_data = StringIO()
export_graphviz(decision_tree, out_file=dot_data, feature_names=predictors,
                class_names=['Unsurvived', 'Survived'], filled=True, rounded=True, special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())
Image(graph.create_png())
```

```python
pred = CART_Reg.predict(X_test) # 模型在测试集上的预测
from sklearn import metrics
metrics.mean_squared_error(y_test, pred) # 计算衡量模型好坏的MSE值
```

## 随机森林模型

### 离散变量

#### 数据准备

```python
import pandas as pd
Titanic = pd.read_csv("C:/Users/ylc/GitHub/Learning/Python/PyhtonDataMiningFromZero/Titanic.csv")
Titanic.head()
```

```python
Titanic.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1, inplace=True) # 删除不使用的变量
Titanic.isnull().sum(axis=0) # 查看缺失值
```

```python
# 对Sex分组，用各组乘客的平均年龄填充各组中的缺失年龄
fillna_Titanic = []
for i in Titanic.Sex.unique():
    update = Titanic.loc[Titanic.Sex==i, ].fillna(value={'Age':Titanic.Age[Titanic.Sex==i].mean()})
    fillna_Titanic.append(update)
Titanic = pd.concat(fillna_Titanic)
# 使用Embarked变量的众数填充缺失值
Titanic.fillna(value={'Embarked':Titanic.Embarked.mode()[0]}, inplace=True)
# 将数值型的Pclass转换为类别型，否则无法对其哑变量处理
Titanic.Pclass = Titanic.Pclass.astype('category')
dummy = pd.get_dummies(Titanic[['Sex', 'Embarked', 'Pclass']]) # 哑变量处理
Titanic = pd.concat([Titanic, dummy], axis=1)
# 删除原始的Sex、Embarked和Pclass变量
Titanic.drop(['Sex', 'Embarked', 'Pclass'], inplace=True, axis=1)
Titanic.head()
```

#### 模型构建

```python
from sklearn import model_selection
predictors = Titanic.columns[1:] # 取出所有自变量名称
# 将数据集拆分为训练集和测试集，且测试集比例为25%
X_train, X_test, y_train, y_test = model_selection.train_test_split(Titanic[predictors], Titanic.Survived,
                                                                    test_size=0.25, random_state=1234)
from sklearn import ensemble
RF_class = ensemble.RandomForestClassifier(n_estimators=200, random_state=1234) # 构建随机森林
RF_class.fit(X_train, y_train) # 随机森林拟合
```

#### 模型预测

```python
RFclass_pred = RF_class.predict(X_test) # m模型在测试集上的预测
from sklearn import metrics
print('模型在测试集的预测准确率：\n', metrics.accuracy_score(y_test, RFclass_pred))
```

#### ROC图

```python
y_score = RF_class.predict_proba(X_test)[:,1]
fpr, tpr, threshold = metrics.roc_curve(y_test, y_score)
roc_auc = metrics.auc(fpr, tpr)
plt.stackplot(fpr, tpr, colors='steelblue', alpha=0.5, edgecolor='black')
plt.plot(fpr, tpr, color='black', lw=1)
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.text(0.5, 0.3, 'ROC curve (area = %0.2f)' % roc_auc)
plt.xlabel('1-Specificity')
plt.ylabel('Sensitivity')
plt.show()
```

#### 变量重要性

```python
importance = RF_class.feature_importances_ # 变量重要性排序
Impt_Series = pd.Series(importance, index=X_train.columns) # 构建含序列用于绘图
Impt_Series.sort_values(ascending=True).plot(kind='barh') # 对序列排序绘图
plt.show()
```
